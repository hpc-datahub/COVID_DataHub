# A python code to make the dictionary

```python
import pandas as pd
import numpy as np
```

```python
base = pd.read_csv("base dictionary.csv", engine='python')
base
```

# Dictionary for cases and deaths

```python
base_cd = base.loc[base["file"] == "casesAndDeaths.csv"]
base_cd
```

```python
dict = {"variable name":[] , "start column": [], "end column":[], "start date":[], "end date": []}
dictionary = pd.DataFrame(dict)
dictionary

df = pd.read_csv("../casesAndDeaths.csv")
name = []
start = []
end = []

start_date = []
end_date = []

file = []

ct = 0

for col in df.columns:
    if col.split("_")[:-1] == []:
        name.append(col)
        start.append(df.columns.get_loc(col))
        start_date.append("")
        file.append("casesAndDeaths.csv")
        if ct != 0:
            end.append(df.columns.get_loc(col) -1) 
            end_date.append("")
    else:       
        new = ("_".join(col.split("_")[:-1])) + "_yyyymmdd"
        
        if new not in name:
            name.append(new)
            start.append(df.columns.get_loc(col))
            start_date.append(col.split("_")[-1])
            file.append("casesAndDeaths.csv")
            if ct != 0:
                end.append(df.columns.get_loc(col) -1) 
                end_date.append(df.columns[df.columns.get_loc(col) -1].split("_")[-1])
    ct = ct+1

end.append(len(df.columns) -1)
end_date.append(df.columns[-1].split("_")[-1])

dictionary["variable name"] = name
dictionary["start column"] = start
dictionary["end column"] = end
dictionary["start date"] = start_date
dictionary["end date"] = end_date

for i in range(dictionary.shape[0]):
    if isinstance(dictionary.iloc[i, 3], str) == True:
        if dictionary.iloc[i, 3].isnumeric() == False:
            dictionary.iloc[i, 3] = ""
    if isinstance(dictionary.iloc[i, 4], str) == True:
        if dictionary.iloc[i, 4].isnumeric() == False:
            dictionary.iloc[i, 4] = ""
dictionary_casesAndDeaths = dictionary
dictionary_casesAndDeaths
```

```python
dictionary_casesAndDeaths = pd.merge(base_cd, dictionary, on="variable name")
dictionary_casesAndDeaths
```

# Dictionary for vaccine data

```python
base_vaccine = base.loc[base["file"] == "vaccine.csv"]
base_vaccine
```

```python
df = pd.read_csv("../vaccine/vaccine.csv")
col_id = [df.columns.get_loc(x) for x in base_vaccine['variable name']]
col_id # find column index according variable name
```

```python
dictionary_vaccine = base_vaccine.copy()
dictionary_vaccine['start column'] = np.array(col_id)
dictionary_vaccine['end column'] = np.array(col_id)
dictionary_vaccine['start date'] = ['20201214'] * 3 + ['20201221'] * 2
dictionary_vaccine['end date'] = ''
dictionary_vaccine
```

```python
# prepare data
d = {}
d['variable name'] = df.columns[-8:]
d['label'] = [x.split('_', 1)[0] + ' updated vaccine ' + x.split('_', 1)[1].replace('_', ' ') \
              for x in d.get('variable name')]
d['source'] = 'https://covid.cdc.gov/covid-data-tracker/#vaccinations'
d['file'] = 'vaccine.csv'
d['start column'] = [df.columns.get_loc(x) for x in d.get('variable name')]
d['end column'] = d.get('start column')
d['start date'] = d.get('variable name')[0].split('_', 1)[0]
d['end date'] = d.get('start date')

# construct 
vac_d = pd.DataFrame(d)
vac_d
```

```python
dictionary_vaccine = dictionary_vaccine.append(vac_d, ignore_index=True)
dictionary_vaccine
```

```python
dictionary = pd.concat([dictionary_casesAndDeaths, dictionary_mobility, dictionary_medical, dictionary_policy, \
                       dictionary_vaccine], sort=False)
dictionary.to_csv("../pandemic_dictionary.csv")
```
