# A python code to make the dictionary

```python
import pandas as pd
import numpy as np
import re
```

```python
base = pd.read_csv("base dictionary.csv", engine='python')
base
```

# Dictionary for cases and deaths

```python
base_cd = base.loc[base["file"] == "casesAndDeaths.csv"]
base_cd
```

```python
dict = {"variable name":[] , "start column": [], "end column":[], "start date":[], "end date": []}
dictionary = pd.DataFrame(dict)
dictionary

df = pd.read_csv("../Pandemic/casesAndDeaths.csv")
name = []
start = []
end = []

start_date = []
end_date = []

file = []

ct = 0

for col in df.columns:
    if col.split("_")[:-1] == []:
        name.append(col)
        start.append(df.columns.get_loc(col))
        start_date.append("")
        file.append("casesAndDeaths.csv")
        if ct != 0:
            end.append(df.columns.get_loc(col) -1) 
            end_date.append("")
    else:       
        new = ("_".join(col.split("_")[:-1])) + "_yyyymmdd"
        
        if new not in name:
            name.append(new)
            start.append(df.columns.get_loc(col))
            start_date.append(col.split("_")[-1])
            file.append("casesAndDeaths.csv")
            if ct != 0:
                end.append(df.columns.get_loc(col) -1) 
                end_date.append(df.columns[df.columns.get_loc(col) -1].split("_")[-1])
    ct = ct+1

end.append(len(df.columns) -1)
end_date.append(df.columns[-1].split("_")[-1])

dictionary["variable name"] = name
dictionary["start column"] = start
dictionary["end column"] = end
dictionary["start date"] = start_date
dictionary["end date"] = end_date

for i in range(dictionary.shape[0]):
    if isinstance(dictionary.iloc[i, 3], str) == True:
        if dictionary.iloc[i, 3].isnumeric() == False:
            dictionary.iloc[i, 3] = ""
    if isinstance(dictionary.iloc[i, 4], str) == True:
        if dictionary.iloc[i, 4].isnumeric() == False:
            dictionary.iloc[i, 4] = ""
dictionary_casesAndDeaths = dictionary
dictionary_casesAndDeaths
```

```python
dictionary_casesAndDeaths = pd.merge(base_cd, dictionary, on="variable name")
dictionary_casesAndDeaths
```

# Dictionary for mobility data

```python
base_mob = base.loc[base["file"].isin(['mobility_google.csv', 'mobility_unacast.csv', 'mobility_descartes.csv'])]
base_mob
```

```python
d = {'variable name':[] , 'start column': [], 'end column':[], 'start date':[], 'end date': []}

for item in ["mobility_google.csv", "mobility_unacast.csv", "mobility_descartes.csv"]:
    df = pd.read_csv("../Pandemic/" + item)
    temp_base = base_mob[base_mob['file'] == item].copy().reset_index(drop = True)
    for i in range(temp_base.shape[0]):
        d['variable name'].append(temp_base.loc[i, 'variable name'])
        temp_var = temp_base.loc[i, 'variable name'].replace('_yyyymmdd', '')
        temp_id = [i for i, s in enumerate(df.columns) if temp_var in s]
        temp_date = [re.findall('20[\d]+', x)[0] for x in df.columns[temp_id]]
        if len(temp_id) > 0:
            d['start column'].append(temp_id[0])
            d['end column'].append(temp_id[-1])
            d['start date'].append(temp_date[0])
            d['end date'].append(temp_date[-1])
        else:
            d['start column'].append(-1)
            d['end column'].append(-1)
            d['start date'].append(-1)
            d['end date'].append(-1)
        
dictionary_mobility = pd.DataFrame(d)
dictionary_mobility
```

```python
# remove the deleted variable in the original data
dictionary_mobility = dictionary_mobility.drop(dictionary_mobility[dictionary_mobility['start column']==-1].index,
                                              axis = 0)
dictionary_mobility
```

```python
dictionary_mobility = pd.merge(base_mob, dictionary_mobility, on="variable name")
dictionary_mobility
```

# Dictionary for vaccine data

```python
base_vaccine = base.loc[base["file"] == "vaccine_allocation.csv"]
base_vaccine
```

```python
df = pd.read_csv("../Pandemic/vaccine_allocation.csv")

d = {'variable name':[] , 'start column': [], 'end column':[], 'start date':[], 'end date': []}

temp_base = base_vaccine.copy().reset_index(drop = True)
for i in range(temp_base.shape[0]):
    d['variable name'].append(temp_base.loc[i, 'variable name'])
    temp_var = temp_base.loc[i, 'variable name'].replace('_yyyymmdd', '')
    temp_id = [i for i, s in enumerate(df.columns) if temp_var in s]
    if temp_var != 'jurisdiction':
        temp_date = [re.findall('20[\d]+', x)[0] for x in df.columns[temp_id]]
    if len(temp_id) > 1:
        d['start column'].append(temp_id[0])
        d['end column'].append(temp_id[-1])
        d['start date'].append(temp_date[0])
        d['end date'].append(temp_date[-1])
    else:
        if len(temp_id) == 1:
            d['start column'].append(temp_id[0])
            d['end column'].append(temp_id[0])
        else:
            d['start column'].append(-1)
            d['end column'].append(-1)
        d['start date'].append(-1)
        d['end date'].append(-1)

dictionary_vaccine = pd.DataFrame(d)
dictionary_vaccine         
```

```python
dictionary_vaccine = pd.merge(base_vaccine, dictionary_vaccine, on="variable name")
dictionary_vaccine
```
