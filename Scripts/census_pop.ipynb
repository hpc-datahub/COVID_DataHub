```python
# importing fips data. and rearranging the columns
fips = pd.read_excel("fips.xlsx")
fips = fips.iloc[:, [1, 3, 2, 4, 5]]

# Convert the 'integers' to integers in fips file.
for i in range(fips.shape[0]):
    if fips.iloc[i, 4] != 'unknown':
        fips.iloc[i, 4] = int(fips.iloc[i, 4])
        fips.iloc[i, 2] = int(fips.iloc[i, 2])

```

```python
# importing the census data

pop = pd.read_excel("US_county_Pop.xlsx")
```

```python
years = [str(item) for item in range(2010,2019)]
params = pop.iloc[:, 6:].columns

col_names = []
for year in years:
    for param in params:
        col_names.append(year + "_" + param)

for col in col_names:
    fips[col] = ""
```

```python
def row_loc_county(inp):
    location = fips.index[(fips['fips']) == inp].values
    location = location.tolist()
    return location
```

```python
for i in range(pop.shape[0]):
    for item in pop.columns[6:]:
        fips.iloc[row_loc_county(pop.iloc[i, 4]), fips.columns.get_loc(str(pop.iloc[i, 5]) + "_" + item)] = pop.iloc[i, pop.columns.get_loc(item)]
        
    
    
```
