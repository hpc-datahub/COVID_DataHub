## GOVEX data

```python
govex = pd.read_csv("https://raw.githubusercontent.com/govex/COVID-19/master/data_tables/restrictions_policies_detailed.csv")
govex.rename(columns={"State code GEOID":"stfips", 'State code STUSPS': 'stabbr', 'State': 'stname'}, inplace=True)
```

```python
govex
```

## BU data

```python
bu = pd.read_csv("~/Documents/ra/HPC_datahub/Raw data/COVID-19 US state policy database 5_5_2021.csv", skiprows=[0, 2, 3, 4], 
                 engine='python')
bu = bu.iloc[:51, :]
bu.rename(columns={'State':'stname', 'State Abbreviation':'stabbr', 'FIPS Code':'stfips'}, inplace=True)
bu['stfips'] = bu['stfips'].astype('int')
bu
```

```python
policy = govex.merge(bu, on = ['stname', 'stfips', 'stabbr'])
policy.to_csv("../Pandemic/policy.csv", index = False)
```
